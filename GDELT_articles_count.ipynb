```python
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv('GDELT_2017/cleaned_jan.csv')
```

```python
df.head()
```

```python
lst = df['13']
```

```python
websites = []
count=0
for l in lst:
    try:
        website = str(l).split('/')[2]
        websites.append(website)
    except:
        count+=1
```

```python
websites = list(set(websites))
```

```python
np.savetxt("websites_jan_gdelt.csv", websites, delimiter=",", fmt='%s')
```

```python
df = pd.read_csv('GDELT_2017/unrest_gdelt_2017_sentiment.csv')
```

```python
df['date'] = ''
for i,row in df.iterrows():
    date = ''
    day = int(row['day'])
    month = int(row['month'])
    year = int(row['year'])
    date = str(day)+'-'+str(month)+'-'+str(year)
    df.set_value(i,'date',date)
```

```python
df.to_csv('GDELT_2017/unrest_gdelt_2017_sentiment.csv')
```
