```python
df = pd.DataFrame(columns= ['Category','Content'])
```

```python
for file in glob.glob(unrest_directory+"*"):
    category = "unrest"
    data = ""
    with open(file,'r') as myfile:
        data = myfile.readlines()[3:]
    data = ''.join(data)
    df = df.append({'Category':category,'Content':data},ignore_index=True)
```
