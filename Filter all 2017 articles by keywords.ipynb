```python
keywords = ['strike',
'unrest',
'masses',
'protest',
'demonstration',
'worker',
'union',
'company',
'caste',
'religious',
'ethnic',
'reformed',
'rebellion',
'defense',
'violence',
'war',
'armed',
'fight',
'Right',
'free',
'freedom',
'liberty',
'justice',
'Fair',
'unfair',
'unequal',
'terror',
'extreme',
'Bomb',
'IED',
'weapon',
'gun',
'wmd',
'threat',
'suicide',
'murder',
'Kill',
'death',
'explosive',
'military',
'police',
'elite',
'government',
'oppresive',
'power',
'regime',
'fraud',
'corruption',
'coup',
'safety',
'secure',
'insecure',
'protect',
'enemy',
'resist',
'hostage',
'truce',
'fire',
'greed',
'panic',
'inflation',
'Price',
'Food',
'Water',
'disorder',
'strife',
'disruption',
'mob',
'tear gas',
'blockade',
'rally',
'sit-in',
'clash',
'barricade',
'roadblock']
```

```python
lmtzr = WordNetLemmatizer()
keywords_lemmatized = []
for k in keywords:
    keywords_lemmatized.append(lmtzr.lemmatize(k))
```

```python
all_data['del'] = ''
```

```python
for i,row in all_data.iterrows():
    if row['category'] == 'unrest':
        all_words = nltk.word_tokenize(str(row['content']))
        all_words_lemmatized = []
        for w in all_words:
            all_words_lemmatized.append(lmtzr.lemmatize(w))
        if len(set(keywords_lemmatized).intersection(set(all_words_lemmatized)))==0:
            all_data.set_value(i,'del',1)
    else:
        all_data.set_value(i,'content',str(row['content']).replace('\\',''))
```

```python
test_data = all_data[all_data['del'] == 1]
test_data.to_csv('GDELT_2017/large_dataset/test_data.csv')
```

```python
all_data = pd.read_csv('GDELT_2017/large_dataset/test_data.csv')
```

```python
stop_words = set(stopwords.words('english')) 
```

```python
for i,row in all_data.iterrows():
    all_words = nltk.word_tokenize(str(row['content']).lower())
    no_stop_words = []
    for w in all_words:
        if w not in stop_words:
            no_stop_words.append(w)
    final_str = " ".join(no_stop_words)
    all_data.set_value(i,'content',final_str)
```

```python
all_data.to_csv('GDELT_2017/large_dataset/test_data_cleaned.csv')
```
